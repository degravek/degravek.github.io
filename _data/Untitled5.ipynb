```python
# Import libraries
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn import cross_validation
from sklearn import preprocessing
from collections import Counter
import matplotlib.pyplot as mp
import seaborn as sns
import pandas as pd
import numpy as np
%matplotlib inline
mp.style.use('ggplot')

# Load the train and test datasets to create two DataFrames
train = pd.read_csv('/Users/degravek/Downloads/train.csv', header=0)
test  = pd.read_csv('/Users/degravek/Downloads/test.csv', header=0)

df = train.append(test,ignore_index=True).reset_index(drop=True)

# Look at the data
ptab = pd.pivot_table(df,index='Sex',columns='Survived',aggfunc='size')
ptab = ptab.div(ptab.sum(axis=1), axis=0)
ptab.plot.barh(stacked=True)
mp.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
mp.xlabel('Fraction of People')
mp.ylabel('Sex')
```
